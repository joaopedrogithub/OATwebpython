```python
#grafico pizza
import pandas as pd
import matplotlib.pyplot as plt
uri = 'https://raw.githubusercontent.com/MedidaSP/nomes-brasileiros-ibge/master/ibge-mas-10000.csv'
pd.read_csv(uri)
nome = pd.read_csv(uri)
nome.head(5)
labels = 'jose', 'joao', 'antonio', 'francisco','carlos'
sizes = [40.1, 20.4, 17.6, 11.9, 10 ]
figl, ax1 = plt.subplots()
plt.title('Os 5 nomes masculinos mais comuns no Brasil')
ax1.pie(sizes,labels=labels, autopct='%1.1f%%', shadow= True, startangle= 90)
ax1.axis('equal')
plt.show()

```
