```python
import pandas as pd 

import matplotlib.pyplot as plt


uri = 'https://raw.githubusercontent.com/adaoduque/Brasileirao_Dataset/master/campeonato-brasileiro-gols.csv'

tabela = pd.read_csv(uri)

display(tabela.head(10))
```

```python
jogador = [] 
gols    = []


n_n = int(input('Quantos Jogadores: '))


for i in range(1, n_n + 1):


  name1 = input('Digite o nome do jogador: ')

  test1 = tabela.loc[tabela['atleta'] == name1.strip()]


  jogador.append(name1)

  gols.append(test1.shape[0])

plt.ylabel('GOLS')
plt.xlabel('ATLETAS')
plt.title('Comparação de Gols')

plt.bar(jogador , gols)
plt.show()
```
